### `PREGUNTA 2:`  
    
Compare visualmente el volumen de venta de los diferentes países:

```python
# Se debe agrupar sumando la cantidad de ventas por país

# Sintaxis de groupby: df.groupby(["lista_columnas_agrupación"])["Columna_valores"].agg(["lista_agregaciones"])

df_global_rpc.groupby(["País"])["Cantidad"].agg("sum")
```

```python
len(df_global_rpc.groupby(["País"]))
```

```python
# Ahora para graficar la agrupación a la sintaxis agregamos el plot con el tipo de gráfico

df_global_rpc.groupby(["País"])["Cantidad"].agg("sum").plot(kind='bar')
```

```python
# Voy a guardar la agrupación anterior en un nuevo DataFrame para poder hacer un Gráfico más personalizado sin inconvenientes
# puesto que si pretendemos por ejemplo, colocar las cantidades sobre las barras, el método plt.annotate buscarará un índice
# numérico, pero en la agrupación se tiene un indice con string que son los países.

df_pais_cantidad = pd.DataFrame(df_global_rpc.groupby(["País"])["Cantidad"].agg("sum"))
df_pais_cantidad["Respaldo País"] = df_pais_cantidad.index
df_pais_cantidad
```

```python
# Coloca un índice numérico dado que en principio eran los países tomados de la agrupación

df_pais_cantidad.index = range(0,4)
df_pais_cantidad
```

```python
# Cambio el nombre del Respaldo País por País y obtengo un DataFrame normal

df_pais_cantidad.rename(columns = {'Respaldo País': 'País'}, inplace = True)
df_pais_cantidad
```

```python
#df_global_rpc.groupby(["País"])["Cantidad"].agg("sum").plot(kind='bar', rot=0)

# Damos formato para una mejor visualización

plt.bar(df_pais_cantidad["País"],df_pais_cantidad["Cantidad"], label='Cantidades')

#Titulo, etiquetas, anotaciones, tamaño, rangos
plt.title("Cantidad de ventas por país")
plt.ylabel("Volumen de ventas")
plt.xlabel("Países")
plt.legend(loc='best')

#Coloco la cantidad sobre cada barra
for i in range(0,4):
    plt.annotate(df_pais_cantidad["Cantidad"][i],(df_pais_cantidad["País"][i],df_pais_cantidad["Cantidad"][i]), size=15, ha='center')

#Dimensiones del gráfico
fig = plt.gcf()
fig.set_size_inches(10.5, 8)

#Establecemos un rango de marcas en el eje Y
#Desde 0 hasta 20 mil en pasos de 1 mil
plt.yticks(ticks=range(0, 20000, 1000))
plt.xticks(rotation=30)
```

### `PREGUNTA 3:`  
    
Verifique la evolución del volumen de ventas anual por cada país. Para ello debe utilizar en el método groupby para agregar a nivel anual la siguiente expresión:

`["País",df_global_rpc["Fecha compra"].map(lambda x: x.year)]`  

Adicionalmente, para que desagregue el gráfico con una línea por país puede usar el método `“unstack”` al final de la sentencia: `.unstack("País").plot()`

```python
# Con df_global_rpc["Fecha compra"].map(lambda x: x.year) obtengo el año de la Fecha Compra

df_global_rpc.groupby(["País",df_global_rpc["Fecha compra"].map(lambda x: x.year)])["Cantidad"].agg("sum")
```

```python
# Grafiquemos la evolución del volumen de ventas por cada país cada año

df_global_rpc.groupby(["País",df_global_rpc["Fecha compra"].map(lambda x: x.year)])["Cantidad"].agg("sum").plot(rot=45)
plt.title("Evolución del volumen de ventas anual por cada país")
plt.ylabel("Volumen de ventas")
```

```python
# Adicionalmente, para que desagregue el gráfico con una línea por país puede usar el método “unstack” 
# al final de la sentencia: .unstack("País").plot()

df_global_rpc.groupby(["País",df_global_rpc["Fecha compra"].map(lambda x: x.year)])["Cantidad"].agg("sum").unstack("País").plot()
plt.title("Evolución del volumen de ventas anual por cada país")
plt.ylabel("Volumen de ventas")

#Dimensiones del gráfico
fig = plt.gcf()
fig.set_size_inches(10.5, 8)

#Establecemos un rango de marcas en el eje Y
#Desde 0 hasta 20 mil en pasos de 1 mil
plt.yticks(ticks=range(0, 6000, 500))
plt.xticks(ticks=range(2018, 2022, 1), rotation=30)
```

```python
# Se observa que ha evolucionado al alza el volumen de ventas por país durante cada año, siendo más estable en Romanía
```
